```python
# Thống kê outliers theo IQR
def count_outliers(series):
    Q1, Q3 = series.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    return ((series < lower) | (series > upper)).sum()

outlier_counts = {col: count_outliers(df[col]) for col in main_features}
print('Số lượng outliers theo IQR:')
for col, count in outlier_counts.items():
    print(f'  {col}: {count:,} ({count/len(df)*100:.2f}%)')
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, silhouette_samples
import warnings
warnings.filterwarnings('ignore')

# Thiết lập style cho biểu đồ
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12
```

```python
df_scaled = pd.read_csv('D:\\MachineLearning\\processed_data\\user_features_scaled.csv')  # Thay đổi đường dẫn file nếu cần

print("=" * 50)
print("THÔNG TIN DỮ LIỆU ĐÃ CHUẨN HÓA")
print("=" * 50)
print(f"Kích thước: {df_scaled.shape}")
print(f"\nCác cột: {df_scaled.columns.tolist()}")
print(f"\nThống kê mô tả:")
df_scaled.describe()
```

```python
# Elbow Method
inertias = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertias.append(kmeans.inertia_)

# Vẽ biểu đồ Elbow
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertias, 'bo-', linewidth=2, markersize=10)
plt.xlabel('Số cụm (K)', fontsize=14)
plt.ylabel('Inertia (Within-cluster Sum of Squares)', fontsize=14)
plt.title('Elbow Method - Tìm số K tối ưu', fontsize=16, fontweight='bold')
plt.xticks(K_range)
plt.grid(True, alpha=0.3)

# Đánh dấu điểm elbow (nếu cần điều chỉnh thủ công)
plt.axvline(x=4, color='r', linestyle='--', label='Điểm Elbow gợi ý (K=4)')
plt.legend()
plt.tight_layout()
plt.savefig('elbow_method.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Biểu đồ Elbow đã được lưu: elbow_method.png")
```

```python
# Silhouette Score cho các giá trị K khác nhau
silhouette_scores = []

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(df_scaled)
    score = silhouette_score(df_scaled, labels)
    silhouette_scores.append(score)
    print(f"K = {k}: Silhouette Score = {score:.4f}")

# Vẽ biểu đồ Silhouette Score
plt.figure(figsize=(10, 6))
plt.plot(K_range, silhouette_scores, 'go-', linewidth=2, markersize=10)
plt.xlabel('Số cụm (K)', fontsize=14)
plt.ylabel('Silhouette Score', fontsize=14)
plt.title('Silhouette Score - Đánh giá chất lượng clustering', fontsize=16, fontweight='bold')
plt.xticks(K_range)
plt.grid(True, alpha=0.3)

# Đánh dấu K tối ưu
optimal_k = K_range[np.argmax(silhouette_scores)]
plt.axvline(x=optimal_k, color='r', linestyle='--', label=f'K tối ưu = {optimal_k}')
plt.legend()
plt.tight_layout()
plt.savefig('silhouette_score.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n🎯 K tối ưu theo Silhouette Score: {optimal_k}")
print("✅ Biểu đồ Silhouette Score đã được lưu: silhouette_score.png")
```

```python
K_OPTIMAL = 4

# Thực hiện K-Means clustering
kmeans_final = KMeans(n_clusters=K_OPTIMAL, random_state=42, n_init=10)
clusters = kmeans_final.fit_predict(df_scaled)

# Thêm nhãn cụm vào dataframe
df_scaled['Cluster'] = clusters

# Thống kê các cụm
print("=" * 50)
print(f"KẾT QUẢ K-MEANS CLUSTERING (K = {K_OPTIMAL})")
print("=" * 50)
print(f"\nSố lượng mẫu trong mỗi cụm:")
print(df_scaled['Cluster'].value_counts().sort_index())

print(f"\nTỷ lệ phần trăm:")
print((df_scaled['Cluster'].value_counts().sort_index() / len(df_scaled) * 100).round(2))

# Silhouette Score cuối cùng
final_silhouette = silhouette_score(df_scaled.drop('Cluster', axis=1), clusters)
print(f"\n📊 Silhouette Score: {final_silhouette:.4f}")
```

```python
# ==============================
# 1. Chuẩn bị dữ liệu (KHÔNG có cột Cluster)
# ==============================
X = df_scaled.drop(columns=['Cluster'], errors='ignore')

# ==============================
# 2. Giảm chiều xuống 5D bằng PCA
# ==============================
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X)

# ==============================
# 3. Train K-means trên dữ liệu GỐC (đã scale)
# ==============================
kmeans_final = KMeans(n_clusters=K_OPTIMAL, random_state=42, n_init=10)
clusters = kmeans_final.fit_predict(X)

# ==============================
# 4. Tạo DataFrame PCA để vẽ
# ==============================
df_pca_result = pd.DataFrame(
    X_pca, 
    columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5']
)
df_pca_result['Cluster'] = clusters

# ==============================
# 5. Vẽ biểu đồ PC1 vs PC2
# ==============================
plt.figure(figsize=(12, 8))
colors = plt.cm.tab10.colors  # Đủ màu cho nhiều cụm

for i in range(K_OPTIMAL):
    cluster_data = df_pca_result[df_pca_result['Cluster'] == i]
    plt.scatter(
        cluster_data['PC1'], 
        cluster_data['PC2'], 
        color=colors[i],
        label=f'Cụm {i}',
        alpha=0.6,
        s=50
    )

# ==============================
# 6. Vẽ centroids (KHÔNG LỖI)
# ==============================
centroids_pca = pca.transform(kmeans_final.cluster_centers_)

plt.scatter(
    centroids_pca[:, 0], 
    centroids_pca[:, 1], 
    color='black', 
    marker='X', 
    s=300, 
    label='Centroids'
)

# ==============================
# 7. Trang trí biểu đồ
# ==============================
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.2f}%)', fontsize=13)
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.2f}%)', fontsize=13)
plt.title('K-Means Clustering - PCA 5D (Visual bằng PC1 & PC2)', fontsize=15, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Lưu ảnh
plt.savefig('kmeans_pca_visualization.png', dpi=300, bbox_inches='tight')
plt.show()

# ==============================
# 8. In thông tin phương sai
# ==============================
print(f"📈 Tổng phương sai giải thích bởi 5 thành phần: {sum(pca.explained_variance_ratio_)*100:.2f}%")
print("✅ Biểu đồ đã được lưu: kmeans_pca_visualization.png")
```

```python
# Phân tích đặc điểm trung bình của mỗi cụm
cluster_analysis = df_scaled.groupby('Cluster').mean()

print("=" * 60)
print("PHÂN TÍCH ĐẶC ĐIỂM CÁC CỤM")
print("=" * 60)
print(cluster_analysis.round(3))

# Vẽ heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(cluster_analysis.T, annot=True, fmt='.2f', cmap='RdYlGn', 
            center=0, linewidths=0.5, cbar_kws={'label': 'Giá trị trung bình (chuẩn hóa)'})
plt.title('Heatmap - Đặc điểm trung bình của các cụm', fontsize=16, fontweight='bold')
plt.xlabel('Cụm', fontsize=14)
plt.ylabel('Đặc trưng', fontsize=14)
plt.tight_layout()
plt.savefig('cluster_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Heatmap đã được lưu: cluster_heatmap.png")
```
